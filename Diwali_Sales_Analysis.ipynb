```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
```

```python
df = pd.read_csv('Diwali_Sales_Data.csv',encoding='unicode_escape')
```

```python
df.shape
```

```python
df.info()
```

```python
# drop unrelevant/blank columns
df.drop(['Status','unnamed1'],axis=1, inplace=True)
```

```python
pd.isnull(df)
```

```python
#check for total null values
pd.isnull(df).sum()
```

```python
#drop null place
df.dropna(inplace=True)
```

```python
#Change Data type
df['Amount'] = df['Amount'].astype('int')
```

# Exploratory Data Analysis

### Gender

```python
ax = sns.countplot(x = 'Gender',data =df)

for bars in ax.containers:
    ax.bar_label(bars)
```

```python
df.groupby(['Gender'], as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False)
```

```python
sales_gen = df.groupby(['Gender'], as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False)
sns.barplot(x = 'Gender', y = 'Amount', data = sales_gen)
```

<i>From above graphs we can see that most of the buyers are females and even the purchasing power of females is greater than men<i>

<h3>Age<h3>

```python
ax = sns.countplot(data= df, x= 'Age Group', hue = 'Gender')

for bars in ax.containers:
    ax.bar_label(bars)
```

```python
#Total Amount vs Age

sales_age = df.groupby(['Age Group'], as_index= False)['Amount'].sum().sort_values(by = 'Amount', ascending= False)
sns.barplot(x = 'Age Group', y= 'Amount', data= sales_age)
```

<i> from above graph we can see that most of the buyers are from age grouop 26-35 and are females <i>

<h3>State<h3>

```python
#total number of orders from top 10 states

sales_state = df.groupby(['State'], as_index=False)['Orders'].sum().sort_values(by='Orders',ascending = False).head(10)

sns.set(rc={'figure.figsize':(15,5)})
sns.barplot(data= sales_state, x= 'State', y= 'Orders')
```

```python
#total amount/sales from top 10 states 

sales_stte = df.groupby(['State'], as_index=False).sum().sort_values(by='Amount', ascending= False).head(10)

sns.set(rc={'figure.figsize':(15,5)})
sns.barplot(data= sales_stte, x= 'State', y= 'Amount')
```

<i> from above we can see that most of the orders and total sales/amount are from Uttar Pradesh, Maharashtra and Karnataka respectively. <i>

<h3> Marital Status <h3>

```python
ax = sns.countplot(data = df, x = 'Marital_Status')

sns.set(rc={'figure.figsize':(2,5)})
for bars in ax.containers:
    ax.bar_label(bars)
```

```python
sales_state = df.groupby(['Marital_Status', 'Gender'],as_index=False).sum().sort_values(by = 'Amount', ascending= False)

sns.set(rc={'figure.figsize':(4,5)})
sns.barplot(data= sales_state, x= 'Marital_Status', y= 'Amount', hue='Gender')
```

<i>from above graphs we can see that most of the buyers are married women and hence they have more purchasing power<i>

### Occupation ###

```python
sns.set(rc={'figure.figsize':(20,7)})
ax= sns.countplot(data = df, x='Occupation')

for bars in ax.containers:
    ax.bar_label(bars)
```

```python
sales_state = df.groupby(['Occupation'],as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False)

sns.set(rc={'figure.figsize':(20,5)})
sns.barplot(data=sales_state, x ='Occupation', y= 'Amount')
```

<i>From above graph we can see that most of the buyers are working from IT,Healthcare and Aviation sectors<i>

### Product Category

```python
sns.set(rc={'figure.figsize':(22,5)})
ax= sns.countplot(data= df, x= 'Product_Category')

for bars in ax.containers:
    ax.bar_label(bars)
```

```python
sales_state = df.groupby(['Product_Category'], as_index=False)['Amount'].sum().sort_values(by='Amount', ascending = False).head(10)

sns.set(rc={'figure.figsize':(22,5)})
sns.barplot(data=sales_state,x='Product_Category',y='Amount')
```

<i> from above graphs we get to know that most of the sold products are from Food, Clothing, Electronics and Footwear category<i>                                                                                           

### Most ordered products

```python
sales_state = df.groupby(['Product_ID'], as_index=False)['Orders'].sum().sort_values(by='Orders', ascending=False).head(10)

sns.set(rc={'figure.figsize':(22,5)})
sns.barplot(data=sales_state, x = 'Product_ID', y = 'Orders')
```

<i> following are the most ordered products with their product ID's<i>
